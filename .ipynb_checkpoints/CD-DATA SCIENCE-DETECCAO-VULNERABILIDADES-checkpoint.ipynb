```python
import requests
from time import sleep
from bs4 import BeautifulSoup
import pandas as pd

import platform
platform.python_version()
```

### **Dataset onde mostra as 20 portas TCP que mais sofreram varreduras em 2020**

```python
#requisição e criação do objeto do tipo BeautifulSoup
response = requests.get('https://www.cert.br/stats/honeypots/')
content = response.content
site = BeautifulSoup(content, 'html.parser')

#procurando a tabela do site
tb = site.find('table', attrs={'class':'stats'})

#convertendo em uma string para pegar via a função 'read_html()'
tb = str(tb)

#Aqui faz de fato, a criação do DataFrame 
tabela = pd.read_html(tb)[0]

#redimensionando o Dataframe para uma melhor visualização
pd.options.display.max_colwidth = 5000

display(tabela)
```
