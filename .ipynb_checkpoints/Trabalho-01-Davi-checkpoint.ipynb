## ***Importações e verificação da versão do Python Kernel.***

```python
import requests
from time import sleep
from bs4 import BeautifulSoup
import pandas as pd

import platform
platform.python_version()
```

## ***Criações das Listas.***

```python
#-------------------------------------------------------------------
sintomas_malwares = []
sintomas_adwares = []
sintomas_spyware = []
sintomas_phishing = []
sintomas_trickbot = []
#-------------------------------------------------------------------
```

## ***Capturando Sintomas de Malwares do Site www.malwarebytes.com.***

```python
#-------------------------------------------------------------------
#limpando a lista de malwares antes de adicionar para garantir que só tenha a quantidade de sintomas corretos, se porventura
#for rodado novamente o codígo e ficar com sintomas duplicados 
sintomas_malwares.clear()


#-------------------------------------------------------------------
#requisição e criação do objeto do tipo BeautifulSoup
response = requests.get('https://www.malwarebytes.com/malware')
content = response.text
site = BeautifulSoup(content, 'html.parser')

#-------------------------------------------------------------------
#Tipo do Malware
tipo_malicius_malware = site.find('h1', attrs={'class':'text-center'}).text

#-------------------------------------------------------------------
#procurei a coluna principal da notícia e coloquei na variável 'coluna_geral' para encontrar as demais informações
coluna_geral = site.find('div', attrs={'class':'col-md-12 col-sm-10 col-restyling'})

#-------------------------------------------------------------------
#Buscando pelos sintomas dentro da principal coluna para capturar os dados para realizar análise se tem algum malware na sua máquina
lista_sintomas = coluna_geral.find('ul', attrs={'type':'disc'})

#-------------------------------------------------------------------
#encontrando os sintomas dentro da coluna geral
todos_sintomas_malware = lista_sintomas.findAll('b')

#-------------------------------------------------------------------
#pegando cada sintoma
sintoma_01_malware = todos_sintomas_malware[0].text
sintoma_02_malware = todos_sintomas_malware[1].text
sintoma_03_malware = todos_sintomas_malware[2].text
sintoma_04_malware = todos_sintomas_malware[3].text
sintoma_05_malware = todos_sintomas_malware[4].text
sintoma_06_malware = todos_sintomas_malware[5].text
sintoma_07_malware = todos_sintomas_malware[6].text
sintoma_08_malware = todos_sintomas_malware[7].text

#-------------------------------------------------------------------

#adicionando na lista
sintomas_malwares.append([tipo_malicius_malware,sintoma_01_malware, sintoma_02_malware, sintoma_03_malware, sintoma_04_malware,sintoma_05_malware,
                         sintoma_06_malware, sintoma_07_malware,sintoma_08_malware])

#-------------------------------------------------------------------
#Display para mostrar os sintomas selecionados para utilizar na categoria de Malware
sintomas_malwares
```

## ***Capturando Sintomas de Adware do Site www.malwarebytes.com - OBSERVAÇÂO: Segue praticamente a mesma estrutura do bloco de códigos anterior.***

```python
#As explicações do código estão no bloco sobre Malware
sintomas_adwares.clear()

#-------------------------------------------------------------------
response2 = requests.get('https://www.malwarebytes.com/adware')
content2 = response2.text
site_adware = BeautifulSoup(content2, 'html.parser')

#-------------------------------------------------------------------

tipo_malicius_adware = site_adware.find('h1', attrs={'class':'text-center'}).text

#-------------------------------------------------------------------

lista_sintomas_adware = site_adware.find('ul', attrs={'type':'disc'})

#-------------------------------------------------------------------

todos_sintomas_adware = lista_sintomas_adware.findAll('li')

#-------------------------------------------------------------------
sintoma_01_adware = todos_sintomas_adware[0].text
sintoma_02_adware = todos_sintomas_adware[1].text
sintoma_03_adware = todos_sintomas_adware[2].text
sintoma_04_adware = todos_sintomas_adware[3].text
sintoma_05_adware = todos_sintomas_adware[4].text
sintoma_06_adware = todos_sintomas_adware[5].text
sintoma_07_adware = todos_sintomas_adware[6].text
sintoma_08_adware = todos_sintomas_adware[7].text

#-------------------------------------------------------------------
sintomas_adwares.append([tipo_malicius_adware,sintoma_01_adware, sintoma_02_adware, sintoma_03_adware, sintoma_04_adware,sintoma_05_adware,
                         sintoma_06_adware, sintoma_07_adware,sintoma_08_adware ])

sintomas_adwares
```

## ***Capturando Sintomas de Spyware do Site www.malwarebytes.com - OBSERVAÇÂO: Segue praticamente a mesma estrutura do bloco de códigos anterior.***

```python
#-------------------------------------------------------------------
sintomas_spyware.clear()

#-------------------------------------------------------------------
response3 = requests.get('https://www.malwarebytes.com/spyware')
content3 = response3.text
site_spyware = BeautifulSoup(content3, 'html.parser')

#-------------------------------------------------------------------

tipo_malicius_spyware = site_spyware.find('h1', attrs={'class':'text-center'}).text

#-------------------------------------------------------------------

lista_sintomas_spyware = site_spyware.find('ul', attrs={'type':'disc'})

#-------------------------------------------------------------------

todos_sintomas_spyware = lista_sintomas_spyware.findAll('strong')

#-------------------------------------------------------------------
sintoma_01_spyware = todos_sintomas_adware[0].text
sintoma_02_spyware = todos_sintomas_adware[2].text
sintoma_03_spyware = todos_sintomas_adware[3].text
sintoma_04_spyware = todos_sintomas_adware[4].text
sintoma_05_spyware = todos_sintomas_adware[6].text
sintoma_06_spyware = todos_sintomas_adware[7].text


#-------------------------------------------------------------------

sintomas_spyware.append([tipo_malicius_spyware,sintoma_01_spyware, sintoma_02_spyware, sintoma_03_spyware, sintoma_04_spyware,sintoma_05_spyware,
                         sintoma_06_spyware, " ", " "])

#-------------------------------------------------------------------
sintomas_spyware
```

## ***Capturando Sintomas de Phishing do Site www.malwarebytes.com - OBSERVAÇÂO: Segue praticamente a mesma estrutura do bloco de códigos anterior.***

```python
#-------------------------------------------------------------------
sintomas_phishing.clear()

#-------------------------------------------------------------------
response4 = requests.get('https://www.malwarebytes.com/phishing')
content4 = response4.text
site_phishing = BeautifulSoup(content4, 'html.parser')

#-------------------------------------------------------------------
# Tratando para pegar apenas o nome do ataque
def formatando_nome(tipo_malicius_phishing):
    tipo_malicius_phishing2 = ""
    
    for letra in tipo_malicius_phishing:
        if letra == '?':
            tipo_malicius_phishing2 += " "
        elif letra == 'p':
            tipo_malicius_phishing2 += "P"
        else:
            tipo_malicius_phishing2 += letra
    return tipo_malicius_phishing2

tipo_malicius_phishing = site_phishing.find('h1', attrs={'class':'text-center'}).text
tipo_malicius_phishing = tipo_malicius_phishing.split()[2]
tipo_malicius_phishing = formatando_nome(tipo_malicius_phishing)

#-------------------------------------------------------------------

lista_sintomas_phishing = site_phishing.find('ul', attrs={'type':'disc'})

#-------------------------------------------------------------------

todos_sintomas_phishing = lista_sintomas_spyware.findAll('li')

#-------------------------------------------------------------------
sintoma_01_phishing = todos_sintomas_phishing[0].text
sintoma_02_phishing = todos_sintomas_phishing[1].text
sintoma_03_phishing = todos_sintomas_phishing[2].text
sintoma_04_phishing = todos_sintomas_phishing[3].text
sintoma_05_phishing = todos_sintomas_phishing[4].text

#-------------------------------------------------------------------

sintomas_phishing.append([tipo_malicius_phishing, sintoma_01_phishing, sintoma_02_phishing, sintoma_03_phishing, sintoma_04_phishing,sintoma_05_phishing,
                         " ", " ", " "])

#-------------------------------------------------------------------
sintomas_phishing
```

## ***Capturando Sintomas de TrickBot do Site www.pcrisk.pt. e do Site https://www.welivesecurity.com.br***

```python
#-----------------------------------------------------------------------------------
#limpando a lista de malwares antes de adicionar para garantir que só tenha a quantidade de sintomas corretos, se porventura
#for rodado novamente o codígo e ficar com sintomas duplicados
sintomas_trickbot.clear()

#------------------------------------------------------------------------------------
#requisição e criação do objeto do tipo BeautifulSoup para o Site pcrisk
response5 = requests.get('https://www.pcrisk.pt/guias-de-remocao/8845-trickbot-virus')
content5 = response5.text
site_trickbot = BeautifulSoup(content5, 'html.parser')

#------------------------------------------------------------------------------------

#pegando toda a tabela para pegar informações úteis para montar o DataSet do primeiro site desta seçção
tabela_trickbot = site_trickbot.findAll('td', attrs={'class':'threat_table_property_value'})
sintoma_01_trickbot = tabela_trickbot[2].text


#------------------------------------------------------------------------------------
#requisição e criação do objeto do tipo BeautifulSoup para o Site welivesecurity
response6 = requests.get('https://www.welivesecurity.com/br/2021/06/25/trickbot-caracteristicas-de-um-dos-malwares-mais-ativos-e-perigosos/')
content6 = response6.text
site_trickbot2 = BeautifulSoup(content6, 'html.parser')


#------------------------------------------------------------------------------------
#coletando o sintoma 02 para inserir no Dataset
coluna_central = site_trickbot2.findAll('p')

#------------------------------------------------------------------------------------
sintoma_02_trickbot = coluna_central[13].text
sintoma_03_trickbot = coluna_central[14].text
sintoma_04_trickbot = coluna_central[16].text
sintoma_05_trickbot = coluna_central[19].text

#------------------------------------------------------------------------------------
#Formatando para pegar apenas o nome do tipo do Malware
tipo_malicius_trickbot = tabela_trickbot[0].text
tipo_malicius_trickbot = tipo_malicius_trickbot.split()[0]
tipo_malicius_trickbot

#------------------------------------------------------------------------------------

sintomas_trickbot.append([tipo_malicius_trickbot, sintoma_01_trickbot, sintoma_02_trickbot, sintoma_03_trickbot, sintoma_04_trickbot, sintoma_05_trickbot, " ", " ", " "])

#------------------------------------------------------------------------------------
#sintomas_trickbot
sintomas_trickbot
```

## ***Criando os DataFrames***

```python
#-------------------------------------------------------------------
#redimensionando o Dataframe para uma melhor visualização
pd.options.display.max_colwidth = 5000


#-------------------------------------------------------------------
#criações dos DataFrames separados para realizar a concatenação em um Dataset único organizado e tratando as informações NaN
data_malwares = pd.DataFrame(sintomas_malwares, index=['Tipo 01'])
data_adwares = pd.DataFrame(sintomas_adwares, index=['Tipo 02'])
data_spyware = pd.DataFrame(sintomas_spyware, index=['Tipo 03'])
data_phishing = pd.DataFrame(sintomas_phishing, index=['Tipo 04'])
data_trickbot = pd.DataFrame(sintomas_trickbot, index=['Tipo 05'])


#-------------------------------------------------------------------
#caso você queira pegar o índice com o nome já definido, a função ignore_index tem que ser 'False'
data_aux_01 = pd.concat((data_malwares,data_adwares),ignore_index=False)
data_aux_02 = pd.concat((data_aux_01,data_spyware))
data_aux_03 = pd.concat((data_aux_02,data_phishing))
data_aux_04 = pd.concat((data_aux_03,data_trickbot))


#-------------------------------------------------------------------
#organizando para realizar as concatenações da melhor maneira possível que encontrei
#data_geral sempre vai receber o resultado da última concatenação
data_geral = data_aux_04
data_geral.columns = ['Software Malicioso','Indicador 1', 'Indicador 2', 'Indicador 3', 'Indicador 4', 'Indicador 5', 'Indicador 6', 'Indicador 7', 'Indicador 8']
data_geral
```

## ***Exportação para um arquivo csv***

```python
data_geral.to_csv('sintomas_softwares_maliciosos.csv', index_label=False)

softwares_maliciosos = pd.read_csv('sintomas_softwares_maliciosos.csv')


#redimensionando o Dataframe para uma melhor visualização
pd.options.display.max_colwidth = 5000
softwares_maliciosos
```
